```python
# 미니탭 반복 입력, 반복수 입력 필수

import pyautogui as ag
import time

noData = input('# of combination :' ) 

wait = 1

time.sleep(wait+4)
for i in range(70, int(noData)+1):
    ag.moveTo(230,62)
    ag.click()
    time.sleep(1)

    # 변수 설정
    ag.moveTo(965,450)
    ag.click()
    ag.press('backspace', presses=21, interval=0.01)
    inputData = 'C'+str((i-1)*7+1)+' C'+str((i-1)*7+3)+' C'+str((i-1)*7+5)
    ag.typewrite(inputData)
    time.sleep(wait)

    # 관측 중단 설정
    ag.moveTo(1180,425)
    ag.click()
    time.sleep(wait)

    ag.moveTo(1040,485)
    ag.click()
    ag.press('backspace', presses=25, interval=0.01)
    inputData2 = 'C'+str((i-1)*7+2)+' C'+str((i-1)*7+4)+' C'+str((i-1)*7+6)
    ag.typewrite(inputData2)
    time.sleep(wait)

    ag.moveTo(1065,525)
    ag.click()
    ag.press('backspace', presses=8)
    ag.typewrite('1')
    time.sleep(wait)

    ag.moveTo(1035,650)
    ag.click()
    time.sleep(wait)

    # 확인
    ag.moveTo(1180,640)
    ag.click()
    time.sleep(wait+4)
```
